```python
import pandas as pd
import nltk
import ssl
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
ssl._create_default_https_context = ssl._create_unverified_context

nltk.download('punkt')

nltk.download([
     "names",
     "stopwords",
     "state_union",
     "twitter_samples",
     "movie_reviews",
     "averaged_perceptron_tagger",
     "vader_lexicon",
     "punkt",
 ])
```

```python
df = pd.read_csv("hotel_reviews 2.csv")
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
from nltk.tokenize import word_tokenize
df['tokens'] = df['Review_Text'].apply(word_tokenize)
```

```python

sia = SentimentIntensityAnalyzer()
def get_sentiment_scores(text):
    return sia.polarity_scores(text)
df['sentiment'] = df['Review_Text'].apply(get_sentiment_scores)
```

```python
df['sentiment']
```

```python

df['compound_score'] = df['sentiment'].apply(lambda x: x['compound'])
df['positive_score'] = df['sentiment'].apply(lambda x: x['pos'])
df['negative_score'] = df['sentiment'].apply(lambda x: x['neg'])
df['neutral_score'] = df['sentiment'].apply(lambda x: x['neu'])

def classify_sentiment(compound_score):
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_category'] = df['compound_score'].apply(classify_sentiment)
```

```python
df.head()
```

```python

def rescale_sentiment(score):
    return (score * 2) - 1

df['rescaled_sentiment_score'] = df['compound_score'].apply(rescale_sentiment)
df['rescaled_positive_score'] = df['positive_score'].apply(rescale_sentiment)
df['rescaled_negative_score'] = df['negative_score'].apply(rescale_sentiment)
df['rescaled_neutral_score'] = df['neutral_score'].apply(rescale_sentiment)

```

```python
df['pos_tags'] = df['tokens'].apply(nltk.pos_tag)
```

```python
topic_keywords = {
    'sanitation': ['clean', 'hygiene', 'sanitation', 'tidy', 'sterile', 'disinfect', 'tidiness', 'sanitize', 'maintained','bathroom'],
    'food': ['food', 'cuisine', 'restaurant', 'dining', 'menu', 'taste', 'delicious', 'culinary','breakfast','tasty', ],
    'hospitality': ['hospitality', 'friendly', 'welcoming', 'attentive', 'service','cosy', 'courteous', 'helpful', 'accommodating', 'comfortable'],
    'value_for_money': ['value for money', 'affordable', 'reasonable', 'cost-effective', 'budget-friendly', 'worth', 'economical', 'inexpensive', 'experience'],
    'facilities': ['facilities', 'amenities', 'services', 'features', 'comforts', 'conveniences', 'offerings', 'provisions','beds','bathroom','attraction','room']
}
```

```python
def calculate_relevance_scores(review):
    scores = {}
    for topic, keywords in topic_keywords.items():
        count = sum(review.lower().count(keyword) for keyword in keywords)
        scores[topic] = count
    return scores
```

```python
df['Relevance_Scores'] = df['Review_Text'].apply(calculate_relevance_scores)

{'sanitation': 1, 'food': 0, 'hospitality': 1, 'value_for_money': 0, 'facilities': 0}
df['sanitation_score'] = df['Relevance_Scores'].apply(lambda x: x['sanitation'])
df['food_score'] = df['Relevance_Scores'].apply(lambda x: x['food'])
df['hospitality_score'] = df['Relevance_Scores'].apply(lambda x: x['hospitality'])
df['value_for_money_score'] = df['Relevance_Scores'].apply(lambda x: x['value_for_money'])
df['facilities_score'] = df['Relevance_Scores'].apply(lambda x: x['facilities'])

# print(list(df['Relevance_Scores']))
```
